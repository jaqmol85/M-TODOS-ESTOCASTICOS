```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios para las variables
n_puntos = 2000
valores_continuos = np.random.uniform(low=0, high=1, size=n_puntos)
valores_discretos = np.random.choice([0, 1], size=n_puntos)

# a. Serie aleatoria
serie_aleatoria = valores_continuos

# b. Proceso de retcula
reticula = np.zeros((10, 10))
for i in range(10):
    for j in range(10):
        reticula[i, j] = valores_continuos[i * 10 + j]

# c. Proceso de tiempo continuo
tiempo_continuo = np.cumsum(valores_continuos)

# d. Proceso de espacio continuo
espacio_continuo = np.linspace(0, 1, n_puntos)

# e. Proceso de espacio-tiempo continuo
espacio_tiempo_continuo = np.outer(valores_continuos, tiempo_continuo)

# f. Proceso de punto compuesto en el tiempo
punto_compuesto_tiempo = np.random.poisson(lam=5, size=n_puntos)

# g. Proceso de punto compuesto en el espacio
punto_compuesto_espacio = np.random.poisson(lam=10, size=n_puntos)

# Visualizacin (opcional)
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.plot(serie_aleatoria)
plt.title("Serie aleatoria")

plt.subplot(2, 3, 2)
plt.imshow(reticula, cmap="viridis", origin="lower")
plt.title("Proceso de retcula")

plt.subplot(2, 3, 3)
plt.plot(tiempo_continuo)
plt.title("Proceso de tiempo continuo")

plt.subplot(2, 3, 4)
plt.plot(espacio_continuo, valores_continuos)
plt.title("Proceso de espacio continuo")

plt.subplot(2, 3, 5)
plt.imshow(espacio_tiempo_continuo, cmap="viridis", origin="lower")
plt.title("Proceso de espacio-tiempo continuo")

plt.subplot(2, 3, 6)
plt.hist(punto_compuesto_tiempo, bins=20, alpha=0.7, color="b", label="Tiempo")
plt.hist(punto_compuesto_espacio, bins=20, alpha=0.7, color="r", label="Espacio")
plt.legend()
plt.title("Proceso de punto compuesto")

plt.tight_layout()
plt.show()
```

a. Serie aleatoria (continua):

Precipitacin diaria: La cantidad de lluvia que cae en un da especfico puede variar continuamente y se puede modelar como una serie aleatoria de valores continuos.
Caudal de un ro: El flujo de agua en un ro puede variar a lo largo del tiempo y se puede representar como una serie continua de valores.
b. Serie aleatoria (discreta):

Nmero de eventos extremos: Por ejemplo, el nmero de inundaciones o sequas en un ao determinado. Estos eventos son discretos y pueden modelarse como una serie aleatoria de valores discretos.
Contaminacin de un ro: La concentracin de contaminantes en un ro puede variar en funcin de eventos puntuales (como descargas industriales) y se puede modelar como una serie discreta de valores.

```python
# Funcin para calcular la semivarianza
def semivarianza(h):
    return 10 * h**1.5

# Desplazamientos (lag) h
desplazamientos = [1, 2, 3, 4, 5]

# Calcular la semivarianza para cada desplazamiento
semivarianzas = [semivarianza(h) for h in desplazamientos]

# Imprimir los resultados
for h, semivar in zip(desplazamientos, semivarianzas):
    print(f"Semivarianza para h = {h}: {semivar:.2f}")
```

Intrnseco: Un proceso es intrnseco si su media no depende del tiempo o del desplazamiento. Dado que la media es constante en este caso, el proceso es intrnseco.
Estacionario en sentido amplio (SSA): Un proceso es SSA si su media y su funcin de covarianza no dependen del tiempo. La semivarianza no afecta la estacionariedad en sentido amplio, por lo que este proceso tambin es SSA.
Estacionario de segundo orden (SSO): Un proceso es SSO si su media es constante y su funcin de covarianza solo depende de la diferencia entre los desplazamientos. La semivarianza no afecta la estacionariedad de segundo orden, por lo que este proceso tambin es SSO.
Estrictamente estacionario: Un proceso es estrictamente estacionario si cumple con las condiciones de estacionariedad de segundo orden y adems su funcin de densidad conjunta no cambia bajo desplazamientos. Dado que la semivarianza no afecta la funcin de densidad conjunta, este proceso tambin es estrictamente estacionario.

6.Demuestre que la escala integral de la funcin de correlacin exponencial es igual al parmetro a y de la funcin de correlacin esfrica es igual a (3/8)a.

6.a.Caso Exponencial

```python
import sympy as sp

# Definir variables simblicas
h, a, s = sp.symbols('h a s', real=True, positive=True)

# Definir la funcin de correlacin esfrica
xi = s**2 * sp.exp(-h/a)

# Calcular la integral de la funcin de correlacin esfrica
integral_xi = sp.integrate(xi, (h, 0, sp.oo))

# Mostrar el resultado


print("El resultado de la integral es:", integral_xi)
```

Este resultado demuestra que la correlacion es igual al parametro a

6. b. Caso Esferica

```python
import sympy as sp

# Definir variables simblicas
h, a, s = sp.symbols('h a s', real=True, positive=True)

# Definir la funcin de correlacin esfrica
xi = s**2 * (1-3/2*(h/a)+1/2*(h/a)**3) * sp.Piecewise((1, h < a), (0, True))

# Calcular la integral de la funcin de correlacin esfrica
integral_xi = sp.integrate(xi, (h, 0, a))

# Mostrar el resultado
print(integral_xi)
```

Con esto se demuestra que la correlacion es igual a 3/8*a

```python
import sympy as sp

# Definimos los smbolos
t, w, tau = sp.symbols('t w tau')

# Definimos la funcin Cz(t)
Cz = sp.Function('Cz')(t)

# Definimos la funcin de densidad espectral Sz(w)
Sz = 1/(2*sp.pi) * sp.integrate(Cz*sp.cos(w*tau), (tau, -sp.oo, sp.oo))

# Ahora, calculamos Cz(0) = desviacion^2 = integral(-inf,inf)Sz(w)dw
desviacion = sp.sqrt(sp.integrate(Sz, (w, -sp.oo, sp.oo)))

Cz0 = desviacion**2

Cz0
```
